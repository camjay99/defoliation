```python
import matplotlib.pyplot as plt
```

```python
doy = [0, 100, 150, 165, 185, 205, 250, 300, 365]
seasonal_trend = [0.1, 0.15, 1, 0.985, 0.965, 0.945, 0.9, 0.15, 0.1]
defoliation = [0.1, 0.15, 1, 0.985, 0.5, 0.945, 0.9, 0.15, 0.1]
logging = [0.1, 0.15, 1, 0.985, 0.5, 0.5, 0.5, 0.15, 0.1]
```

```python
plt.plot(doy, seasonal_trend, '-', color='black', label='Seasonal Trend')
plt.plot(doy, defoliation, '--', color = 'blue', label='Defoliation')
plt.plot(doy, logging, '-.', color = 'red', label = 'Logging')
plt.legend()
plt.ylabel('Normalized EVI', fontsize=15)
plt.xlabel('Day of Year', fontsize=15)
plt.savefig('../data/hyptrends.png')
```
