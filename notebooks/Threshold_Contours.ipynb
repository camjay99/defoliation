```python
import proplot as pplt
import matplotlib as plt
import numpy as np
import pandas as pd
import rasterio as rio
```

## Allegheny

```python
src = rio.open('./Data/Allegheny_Threshold_Contours.tif')
allegheny = src.read()
allegheny = allegheny.astype(float)
allegheny[allegheny == 0] = np.nan
```

```python
fig, ax = pplt.subplots(figsize=(8, 8))

ax.imshow(allegheny[0], c='dark red')
ax.scatter([], [], c='dark red',  label='Threshold = 0.1') # Invisible artist for adding to legend

ax.imshow(allegheny[10], c='dark yellow')
ax.scatter([], [], c='dark yellow',  label='Threshold = 0.2') # Invisible artist for adding to legend

ax.format(xlim=[1500, 4500], ylim=[5000, 2000], xticks=[], yticks=[])
pplt.rc['font.size'] = 20
ax.legend(loc='lr', ncols=1)

fig.savefig('./Figures/Contours/Allegheny_Countours.svg')
```
