```python
import proplot as pplt
import matplotlib as plt
import numpy as np
import pandas as pd
```

```python
landcover_counts = pd.read_csv('./Data/Enviro/landcover_counts.csv')
landcover_counts['rate'] = landcover_counts['intense'] / landcover_counts['total']
```

```python
fig, axes = pplt.subplots(nrows=2, ncols=2, share=1)

counts_2019 = landcover_counts[landcover_counts['year'] == 2019.0]
axes[0].bar(counts_2019['landcover'], counts_2019['rate'])
axes[0].format(xlocator=[0, 1, 2], xminorlocator=0.5, ytickminor=False, xticklabels=['Deciduous', 'Conifer', 'mixed'], ylabel='% Area Defoliated', xlabel='Landcover', labelsize=12, title='2019')

counts_2020 = landcover_counts[landcover_counts['year'] == 2020.0]
axes[1].bar(counts_2020['landcover'], counts_2020['rate'])
axes[1].format(xlocator=[0, 1, 2], xminorlocator=0.5, ytickminor=False, xticklabels=['Deciduous', 'Conifer', 'Mixed'], ylabel='% Area Defoliated', xlabel='Landcover', labelsize=12, title='2020')

counts_2021 = landcover_counts[landcover_counts['year'] == 2021.0]
axes[2].bar(counts_2021['landcover'], counts_2021['rate'])
axes[2].format(xlocator=[0, 1, 2], xminorlocator=0.5, ytickminor=False, xticklabels=['Deciduous', 'Conifer', 'Mixed'], ylabel='% Area Defoliated', xlabel='Landcover', labelsize=12, title='2021')

counts_2022 = landcover_counts[landcover_counts['year'] == 2022.0]
axes[3].bar(counts_2022['landcover'], counts_2022['rate'])
axes[3].format(xlocator=[0, 1, 2], xminorlocator=0.5, ytickminor=False, xticklabels=['Deciduous', 'Conifer', 'Mixed'], ylabel='% Area Defoliated', xlabel='Landcover', labelsize=12, title='2022')

#fig.savefig('./Figures/landcover_defoliation_rate.png')
```

```python
fig, ax = pplt.subplots(figsize=('90mm','90mm'), fontsize=10)

linewidth = 5

deciduous = landcover_counts[landcover_counts['landcover'] == 'deciduous'].reset_index()
ax.plot(deciduous['year'], deciduous['rate'], linewidth=linewidth, color='green', label='Deciduous')

mixed = landcover_counts[landcover_counts['landcover'] == 'mixed'].reset_index()
ax.plot(mixed['year'], mixed['rate'], linewidth=linewidth, color='blue', label='Mixed')

evergreen = landcover_counts[landcover_counts['landcover'] == 'evergreen'].reset_index()
ax.plot(evergreen['year'], evergreen['rate'], linewidth=linewidth, color='orange', label='Evergreen')

ax.format(xlabel='Year', ylabel='% Landcover Area Defoliated', xlim=[2020, 2023], xlocator=[2020, 2021, 2022, 2023], xminorlocator=[])
ax.legend(ncols=1)

fig.savefig('./Figures/landcover_defoliation_rate.png')
```

```python
def calc_prop(x):
    x['proportion'] = x['intense']/x['intense'].sum()
    x['land_prop'] = x['total']/x['total'].sum()
    return x

landcover_counts = landcover_counts.groupby(by='year', group_keys=False).apply(calc_prop)
```

```python
fig, ax = pplt.subplots(figsize=('90mm','90mm'), fontsize=10)

linewidth = 5

deciduous = landcover_counts[landcover_counts['landcover'] == 'deciduous'].reset_index()
ax.plot(deciduous['year'], deciduous['proportion'], linewidth=linewidth, color='green', label='Deciduous')
ax.plot(deciduous['year'], deciduous['land_prop'], linewidth=linewidth, color='green', linestyle='dashed', label='')#, label='Deciduous')


mixed = landcover_counts[landcover_counts['landcover'] == 'mixed'].reset_index()
ax.plot(mixed['year'], mixed['proportion'], linewidth=linewidth, color='blue', label='Mixed')
ax.plot(mixed['year'], mixed['land_prop'], linewidth=linewidth, color='blue', linestyle='dashed', label='')

evergreen = landcover_counts[landcover_counts['landcover'] == 'evergreen'].reset_index()
ax.plot(evergreen['year'], evergreen['proportion'], linewidth=linewidth, color='orange', label='Evergreen')
ax.plot(evergreen['year'], evergreen['land_prop'], linewidth=linewidth, color='orange', linestyle='dashed', label='')

ax.format(xlabel='Year', ylabel='% of Defoliated Area', xlim=[2020, 2023], xlocator=[2019, 2020, 2021, 2022, 2023], xminorlocator=[], ylim=[0, 1])
ax.legend(ncols=1, loc='ur')

fig.savefig('./Figures/landcover_defoliation_proportion.png')
```

```python
landcovers = ['Deciduous', 'Mixed', 'Evergreen']
defol_props = np.array([deciduous['proportion'].mean(),
               mixed['proportion'].mean(),
               evergreen['proportion'].mean()])
land_props = np.array([deciduous['land_prop'].mean(),
               mixed['land_prop'].mean(),
               evergreen['land_prop'].mean()])
```

```python
fig, ax = pplt.subplots(figsize=('120mm','90mm'), fontsize=10)

ax.bar(landcovers, defol_props*100, color=['green', 'blue', 'orange'])

ax.hlines(land_props*100, [-0.4, 0.6, 1.6], [0.4, 1.4, 2.4], color='black', linestyle='solid', linewidth=3, label='% of All Forest')

ax.format(xlocator=[0, 1, 2], xminorlocator=0.5, ytickminor=False, xticklabels=['Deciduous', 'Mixed', 'Conifer'], 
          ylim=[0, 75], ylabel='% of Defoliated Regions', xlabel='Landcover', labelsize=12)

ax.legend(loc='r')

fig.savefig('./Figures/Enviro/landscape_bar_plots.png')
```
