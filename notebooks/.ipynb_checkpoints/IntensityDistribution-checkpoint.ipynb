# Intensity Distribution
Here, we plot histograms of mean intensity derived from GEE.

```python
import numpy as np
import pandas as pd
import ultraplot as uplt
```

## Helper Functions

```python
def gather_subtokens(tokens):
    subtokens = []
    for token in tokens:
        subtokens.append(token)
        if ']' in token:
            break
    return subtokens

def gather_entries(tokens):
    entries = []
    i = 0
    while i < len(tokens):
        if '=' in tokens[i]:
            entries.append(tokens[i].split('='))
            if '[' in entries[-1][1]:
                # Gather subtokens in the list
                subtokens = gather_subtokens(tokens[i+1:])
                entries[-1][1] = [entries[-1][1]] + subtokens
                # Remove leading and trailing brackets
                entries[-1][1][0] = entries[-1][1][0].strip('[')
                entries[-1][1][-1] = entries[-1][1][-1].strip(']')
                i += len(subtokens)
        i += 1
    return entries
```

## Load Data

```python
data = pd.read_csv('./Data/defol_intensity_2020.csv')
tokens_2020 = data.iloc[0,0].strip('{}').split(', ')
entries_2020 = gather_entries(tokens_2020)

# Put values in python variables
bucket_min_2020 = entries_2020[0][1]
bucket_width_2020 = entries_2020[1][1]
histogram_2020 = np.array(list(map(float, entries_2020[2][1])))
bucket_means_2020 = np.array(list(map(float, entries_2020[3][1])))
```

```python
data = pd.read_csv('./Data/defol_intensity_2021.csv')
tokens_2021 = data.iloc[0,0].strip('{}').split(', ')
entries_2021 = gather_entries(tokens_2021)

# Put values in python variables
bucket_min_2021 = entries_2021[0][1]
bucket_width_2021 = entries_2021[1][1]
histogram_2021 = np.array(list(map(float, entries_2021[2][1])))
bucket_means_2021 = np.array(list(map(float, entries_2021[3][1])))
```

```python
data = pd.read_csv('./Data/defol_intensity_2022.csv')
tokens_2022 = data.iloc[0,0].strip('{}').split(', ')
entries_2022 = gather_entries(tokens_2022)

# Put values in python variables
bucket_min_2022 = entries_2022[0][1]
bucket_width_2022 = entries_2022[1][1]
histogram_2022 = np.array(list(map(float, entries_2022[2][1])))
bucket_means_2022 = np.array(list(map(float, entries_2022[3][1])))
```

```python
data = pd.read_csv('./Data/defol_intensity_2023.csv')
tokens_2023 = data.iloc[0,0].strip('{}').split(', ')
entries_2023 = gather_entries(tokens_2023)

# Put values in python variables
bucket_min_2023 = entries_2023[0][1]
bucket_width_2023 = entries_2023[1][1]
histogram_2023 = np.array(list(map(float, entries_2023[2][1])))
bucket_means_2023 = np.array(list(map(float, entries_2023[3][1])))
```

## Plot Histogram

```python
# Convert to hectares, each pixel is 100 m^2 and 10,000 m^2 in hectare
histogram_2020 = histogram_2020*100/10_000
histogram_2021 = histogram_2021*100/10_000
histogram_2022 = histogram_2022*100/10_000
histogram_2023 = histogram_2023*100/10_000
```

```python
fig, axes = uplt.subplots(nrows=2, ncols=2, fontsize=12)

axes.format(abc='A.', abcloc='ul')

axes.format(ylim=[0.5,1_000_000])

mask = bucket_means_2020 <= -0.04
axes[0].bar(bucket_means_2020[mask], histogram_2020[mask], width=1, facecolor='blood red')
axes[0].bar(bucket_means_2020[~mask], histogram_2020[~mask], width=1, facecolor='cornflower')
axes[0].format(title='2020', xlabel='Mean Intensity', ylabel='Area (ha)', yscale='log')

mask = bucket_means_2021 <= -0.04
axes[1].bar(bucket_means_2021[mask], histogram_2021[mask], width=1, facecolor='blood red')
axes[1].bar(bucket_means_2021[~mask], histogram_2021[~mask], width=1, facecolor='cornflower')
axes[1].format(title='2021', xlabel='Mean Intensity', ylabel='Area (ha)', yscale='log')

mask = bucket_means_2022 <= -0.04
axes[2].bar(bucket_means_2022[mask], histogram_2022[mask], width=1, facecolor='blood red')
axes[2].bar(bucket_means_2022[~mask], histogram_2022[~mask], width=1, facecolor='cornflower')
axes[2].format(title='2022', xlabel='Mean Intensity', ylabel='Area (ha)', yscale='log')

mask = bucket_means_2023 <= -0.04
axes[3].bar(bucket_means_2023[mask], histogram_2023[mask], width=1, facecolor='blood red')
axes[3].bar(bucket_means_2023[~mask], histogram_2023[~mask], width=1, facecolor='cornflower')
axes[3].format(title='2023', xlabel='Mean Intensity', ylabel='Area (ha)', yscale='log')

fig.savefig('./Figures/IntensityDistribution/log_intensity.png')
```

```python
fig, axes = uplt.subplots(nrows=2, ncols=2, fontsize=12)

axes.format(abc='A.', abcloc='ul')

axes.format(ylim=[0.5,650_000])

mask = bucket_means_2020 <= -0.04
axes[0].bar(bucket_means_2020[mask], histogram_2020[mask], width=1, facecolor='blood red')
axes[0].bar(bucket_means_2020[~mask], histogram_2020[~mask], width=1, facecolor='cornflower')
axes[0].format(title='2020', xlabel='Mean Intensity', ylabel='Area (ha)')

mask = bucket_means_2021 <= -0.04
axes[1].bar(bucket_means_2021[mask], histogram_2021[mask], width=1, facecolor='blood red')
axes[1].bar(bucket_means_2021[~mask], histogram_2021[~mask], width=1, facecolor='cornflower')
axes[1].format(title='2021', xlabel='Mean Intensity', ylabel='Area (ha)')

mask = bucket_means_2022 <= -0.04
axes[2].bar(bucket_means_2022[mask], histogram_2022[mask], width=1, facecolor='blood red')
axes[2].bar(bucket_means_2022[~mask], histogram_2022[~mask], width=1, facecolor='cornflower')
axes[2].format(title='2022', xlabel='Mean Intensity', ylabel='Area (ha)')

mask = bucket_means_2023 <= -0.04
axes[3].bar(bucket_means_2023[mask], histogram_2023[mask], width=1, facecolor='blood red')
axes[3].bar(bucket_means_2023[~mask], histogram_2023[~mask], width=1, facecolor='cornflower')
axes[3].format(title='2023', xlabel='Mean Intensity', ylabel='Area (ha)')

fig.savefig('./Figures/IntensityDistribution/intensity.png')
```

```python
resample_2021 = np.interp(bucket_means_2020, bucket_means_2021, histogram_2021, 0, 0)
resample_2022 = np.interp(bucket_means_2020, bucket_means_2022, histogram_2022, 0, 0)
resample_2023 = np.interp(bucket_means_2020, bucket_means_2023, histogram_2023, 0, 0)
histogram_cumulative = histogram_2020 + resample_2021 + resample_2022 + resample_2023
```

```python
fig, ax = uplt.subplots(fontsize=12)

ax.bar(bucket_means_2020, histogram_cumulative, width=1, facecolor='blood red')
ax.format(title='Cumulative', xlabel='Mean Intensity', ylabel='Area Impacted (ha)')

fig.savefig('./Figures/IntensityDistribution/cumulative_intensity.png')
```
