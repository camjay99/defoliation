```python
# import GEE-related libraries
import ee

try:
    ee.Initialize()
except:
    # need to authenticate with your credential at the first time
    ee.Authenticate()
    ee.Initialize()
#import geemap 
import geemap.eefolium as geemap

# import other necessary libraries
from calendar import monthrange
import numpy as np
import rasterio
import pandas as pd
from pathlib import Path
import xarray as xr
from matplotlib import pyplot as plt
%matplotlib inline 
from pygam import LinearGAM,s
from datetime import datetime
from pygam import ExpectileGAM 
```

```python
# Spatial range of Mount Pleasant sites
poly = ee.Geometry.Polygon(
       [[-76.383, 42.47],
        [-76.387, 42.47],
        [-76.387, 42.465],
        [-76.383, 42.465]])
```

```python
# cloud and shadow mask
def maskCloudAndShadows(image): 
    # cloud probability less than 20%
    cloudProb = image.select('MSK_CLDPRB')
    cloud = cloudProb.lt(20)
    
    # shadow, cirrus, high cloud probability: based on SCL band
    scl = image.select('SCL')
    shadow = scl.eq(3)
    cirrus = scl.eq(10)
    cloud_high = scl.eq(9)
    
    # cirrus and cloud: based on QA60 band
    qa = image.select('QA60')
    cloudBitMask = 1 << 10
    cirrusBitMask = 1 << 11
    mask1 = (qa.bitwiseAnd(cloudBitMask).eq(0))
    mask2 = (qa.bitwiseAnd(cirrusBitMask).eq(0))
    qa_mask = mask1.bitwiseAnd(mask2)
    
    mask = ((cloud.eq(1)).bitwiseAnd(cirrus.neq(1)).bitwiseAnd(shadow.neq(1)).bitwiseAnd(cloud_high.neq(1)).bitwiseAnd(qa_mask))
    return image.select(['B2','B4','B8']).updateMask(mask).divide(10000)
   
```

```python
# first get unique date
imcol = (
    ee.ImageCollection('COPERNICUS/S2_SR')
            .filterDate('2019-11-01'
                        ,'2021-11-04')
            .filterBounds(poly)
            .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE',50))
            .map(lambda img: img.set('date', img.date().format('YYYY-MM-dd'))) 
)

date_range = imcol.reduceColumns(ee.Reducer.toList(1), ["date"]).values().get(0).getInfo()
# convert to datetime
date_range = np.unique(np.squeeze(date_range)).astype('datetime64')
# add 1 day to get end date
date_end = date_range + np.timedelta64(1,'D')
```

```python
# then loop over range to download data
for date, date_end in zip(date_range.astype('str'),date_end.astype('str')):
    fn = f'mtpleas_{date}.tif'
    
    image = ( 
        ee.ImageCollection('COPERNICUS/S2_SR')
                .filterDate(date,date_end)
                .filterBounds(poly) 
                .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE',50))
                .map(maskCloudAndShadows)
                .mosaic()
                .clip(poly.bounds())
                )

    geemap.ee_export_image(image, f'./{fn}',scale=10)
    
```

```python
# calculate Red, Grn, Blu, Nir, EVI, NIRv, and NDWI for each month
# combine them into a csv file
# create another data file to record the lat and lon for each grid cell

evi_func = lambda nir, red, blu: 2.5 * (nir - red) / (nir + 6 * red - 7.5 * blu + 1.)
ndvi_func = lambda nir, red: (nir - red) / (nir + red)
nirv_func = lambda nir, red: nir * (nir - red) / (nir + red)

# get the filenames for all tif files
data_path = Path('/home/qw64/Mt_Pleasant_GEE/') 
file_list = np.array(sorted(list(data_path.glob('*.tif'))))

# an open list to store data array
da_list = []

for filename in file_list:
    date = filename.name.split('_')[-1][0:-4] # get rid of .tif
    with rasterio.open(filename, 'r') as src:
        im = src.read()
        # read image into ndarray

        # transpose the matrix from (band, row, col) to (row, col, band)
        im = np.transpose(im, [1,2,0])

        # set null values
        im[im == 0.] = np.nan

        # get EVI, NDVI, NIRv
        EVI = evi_func(im[:,:,2],im[:,:,1],im[:,:,0])
        NDVI = ndvi_func(im[:,:,2],im[:,:,1])
        NIRv = nirv_func(im[:,:,2],im[:,:,1])

        # get coordinates
        # first axis, from north to south
        lat_array = np.linspace(src.bounds[3],src.bounds[1],im.shape[0])
        # second axis, from west to east
        lon_array = np.linspace(src.bounds[0],src.bounds[2],im.shape[1])

        src.close()
        
        da = xr.Dataset({'EVI':(('lat','lon'),EVI), 'NDVI':(('lat','lon'),NDVI), 'NIRv':(('lat','lon'),NIRv)},
                        coords={'lat':lat_array,'lon':lon_array})
        
        # add time dimension
        da = da.expand_dims('time')
        da.coords['time'] = [np.datetime64(date)]
        
        # append the data array to da_list
        
        da_list.append(da)
```

```python
# for each VI, pixel, and year combination, we fit a expectile GAM model to smooth the data (reduce the noise)
# the idea is similar to quantile regression
# k controls the quantile

def vi_diff_expectile(i,j,k,data):
    # generate the year masks
    _vi = data.vi[:,i,j].values
    _doy = data.doy.values
    _year = data.year.values
    _year2019 = [int(i)==2019 for i in _year]
    _year2020 = [int(i)==2020 for i in _year]
    _year2021 = [int(i)==2021 for i in _year]
    _mask = np.isfinite(_vi)

    df_gam_2019 = pd.DataFrame({'VI' : _vi[_mask & _year2019], 
                                'DOY' : _doy[_mask & _year2019]})
    df_gam_2020 = pd.DataFrame({'VI' : _vi[_mask & _year2020], 
                                'DOY' : _doy[_mask & _year2020]})
    df_gam_2021 = pd.DataFrame({'VI' : _vi[_mask & _year2021], 
                                'DOY' : _doy[_mask & _year2021]})
    df_gam_pre=df_gam_2019.append(df_gam_2020)

    if df_gam_2021.empty:
        return(float('nan'))
    else: 
        XX_pre = np.arange(1,np.max(df_gam_pre['DOY'])+1)
        XX_2021 = np.arange(1,np.max(df_gam_2021['DOY'])+1)
    
        # model
        gam50_pre = ExpectileGAM(expectile=0.5).gridsearch(df_gam_pre['DOY'][:,None],df_gam_pre['VI'])
        lam_pre = gam50_pre.lam
        gam_pre_95=ExpectileGAM(n_splines=8,expectile=k, lam=lam_pre).fit(df_gam_pre['DOY'],df_gam_pre['VI'])
    
        gam50_2021 = ExpectileGAM(expectile=0.5).gridsearch(df_gam_2021['DOY'][:,None],df_gam_2021['VI'])
        lam_2021 = gam50_2021.lam
        gam_2021_95=ExpectileGAM(n_splines=5,expectile=k, lam=lam_2021).fit(df_gam_2021['DOY'],df_gam_2021['VI'])
    
        # smoothed data
        pred_pre = pd.DataFrame({'DOY': XX_pre,'predicted VI': gam_pre_95.predict(XX_pre)})
        pred_2021 = pd.DataFrame({'DOY': XX_2021,'predicted VI': gam_2021_95.predict(XX_2021)})

        pred_2021_july=np.mean(pred_2021.query('DOY >= 182 & DOY <= 213')['predicted VI'])
        pred_pre_july=np.mean(pred_pre.query('DOY >= 182 & DOY <= 213')['predicted VI'])
        
        
        pred_2021_sept=np.mean(pred_2021.query('DOY >= 244 & DOY <= 274')['predicted VI'])
        pred_pre_sept=np.mean(pred_pre.query('DOY >= 244 & DOY <= 274')['predicted VI'])
        
        
        
        #pred_pre_winter=np.mean(pred_pre.query('DOY >= 305 or DOY <= 60')['predicted VI'])
        change=(pred_2021_july-pred_pre_july)/pred_pre_july*100
        return([pred_pre_july,pred_2021_july,pred_pre_sept,pred_2021_sept,change]) # 2 more ouputs
    
```

```python
# Loop over each of the three vegetation indices to save 1)spatial pattern of VI difference; 2)seasonal trend at AS-1 site.

for this_vi in ['EVI']:
    ds_vi_mtpleas = ds_mtpleas.where((ds_mtpleas[this_vi] < 1.) & (ds_mtpleas[this_vi] > 0.),drop=True)
    ds_vi_mtpleas = ds_vi_mtpleas.rename({this_vi:'vi'})
    
    # spatial pattern of VI difference
    
    # generate the data frame that stores the VI difference (calculated based on expectile GAM)
    lat_july=[]
    lat_july_index=[]
    lon_july=[]
    lon_july_index=[]
    vi_july_pre=[]
    vi_july_2021=[]
    vi_sept_pre=[]
    vi_sept_2021=[]
    vi_july_change=[]
    #vi_winter=[]
    for i in range(47): #latitude
        for j in range(28): #longitude
            lat_july.append(float(ds_vi_mtpleas['lat'][i]))
            lat_july_index.append(i)
            lon_july.append(float(ds_vi_mtpleas['lon'][j]))
            lon_july_index.append(j)
            temp=vi_diff_expectile(i,j,0.9,ds_vi_mtpleas)
            vi_july_pre.append(temp[0])
            vi_july_2021.append(temp[1])
            vi_sept_pre.append(temp[2])
            vi_sept_2021.append(temp[3])
            vi_july_change.append(temp[4])
            #vi_winter.append(temp[3])

    vi_diff_july_expectile=pd.DataFrame({'Latitude' : lat_july, 
                               'Longitude' : lon_july, 
                               'lat_index' : lat_july_index,
                               'lon_index' : lon_july_index,
                               'VI in July (2021)' : vi_july_2021,
                               'VI in July (2019-2020)' : vi_july_pre,
                               'VI in Sept (2021)' : vi_sept_2021,
                               'VI in Sept (2019-2020)' : vi_sept_pre,         
                               'VI difference' : vi_july_change})
    vi_diff_july_expectile.to_csv(f'/home/qw64/Mt_Pleasant_GEE/mtpleas_{this_vi}_diff_july_sept_expectile_10_27.csv')

    
    # find the coordinates of the AS-1 plot
    as1_mtpleas=ds_vi_mtpleas.where((ds_vi_mtpleas['lat']<=42.46874) & 
                                    (ds_vi_mtpleas['lat']>=42.46832) & 
                                    (ds_vi_mtpleas['lon']>=-76.38577) & 
                                    (ds_vi_mtpleas['lon']<=-76.38535),drop=True)
    _vi = as1_mtpleas.vi.values
    _lat = as1_mtpleas.lat.values
    _lon = as1_mtpleas.lon.values

    # loop over the coordinates
    # record 1)DOY-VI; 2)whether the data comes from a negatively-affected pixel; 3)whether the data comes from 2021 or prior years
    as1_doy=[]
    as1_vi=[]
    label_damage=[]
    label_year=[]
    for thislat in _lat:
        for thislon in _lon:
            lat_index=pd.Index(ds_vi_mtpleas['lat'].values).get_loc(thislat)
            lon_index=pd.Index(ds_vi_mtpleas['lon'].values).get_loc(thislon)
            _vi = ds_vi_mtpleas.vi[:,lat_index,lon_index].values
            _doy = ds_vi_mtpleas.doy.values
            _year = np.array([int(i)==2021 for i in ds_vi_mtpleas.year.values])
            _mask = [(i>0) & (i<1) for i in _vi]
            
            # label_year: if true, then this is the data from 2021; if false, then it's from previous years
            label_year.extend(_year[_mask].astype(int).astype(str))
            as1_doy.extend(_doy[_mask])
            as1_vi.extend(_vi[_mask])
            
            n=len(_doy[_mask])
            
            # label the data: whether it has reduced VI in July or not
            #check_damage=vi_diff_expectile(lat_index,lon_index,0.9,ds_vi_mtpleas)[2]
            check_damage=vi_diff_july_expectile.query('Latitude==@thislat & Longitude==@thislon')['VI difference'].values
            if check_damage>0:
                label_damage.extend(['Increase']*n)
            else:
                label_damage.extend(['Decrease']*n)
                
    as1_data = pd.DataFrame({'VI':as1_vi, 'DOY':as1_doy, 'year':label_year,'damage':label_damage})
    
    # for each year and damage combination, we fit a GAM model
    
    # one data frame for observed seasonal trend
    as1_obs_vi_gam=[]
    as1_obs_doy_gam=[]
    as1_obs_year_gam=[]
    as1_obs_damage_gam=[]
    # one data frame for GAM-fitted seasonal trend
    as1_pred_vi_gam=[]
    as1_pred_doy_gam=[]
    as1_pred_year_gam=[]
    as1_pred_damage_gam=[]
    
    # loop over each year*damage combination
    for thisyear in ['1','0']:
        for thisdamage in ['Increase','Decrease']:
            tempdata=as1_data.query('year==@thisyear & damage==@thisdamage')
            gam_search = ExpectileGAM(expectile=0.5).gridsearch(tempdata['DOY'][:,None],tempdata['VI'])
            thislam = gam_search.lam
            gam_90 = ExpectileGAM(n_splines=8,expectile=0.9, lam=thislam).fit(tempdata['DOY'][:,None],tempdata['VI'])
            xx_pred = np.arange(1,np.max(tempdata['DOY'])+1)
            pred=gam_90.predict(xx_pred)
            xx_obs=tempdata.groupby(by='DOY').mean().index
            obs=tempdata.groupby(by='DOY').mean()
            
            as1_obs_vi_gam.extend(obs['VI'])
            as1_obs_doy_gam.extend(xx_obs)
            as1_obs_year_gam.extend(thisyear*obs.shape[0])
            as1_obs_damage_gam.extend([thisdamage]*obs.shape[0])
            
            as1_pred_vi_gam.extend(pred)
            as1_pred_doy_gam.extend(xx_pred)
            as1_pred_year_gam.extend(thisyear*len(pred))
            as1_pred_damage_gam.extend([thisdamage]*len(pred))
    
    obs=pd.DataFrame({'VI':as1_obs_vi_gam, 'DOY':as1_obs_doy_gam, 'year':as1_obs_year_gam,'damage':as1_obs_damage_gam})
    pred=pd.DataFrame({'VI':as1_pred_vi_gam, 'DOY':as1_pred_doy_gam, 'year':as1_pred_year_gam,'damage':as1_pred_damage_gam})
    
    obs['year']=['2021' if i=='1' else '2019-2020' for i in obs['year']]
    pred['year']=['2021' if i=='1' else '2019-2020' for i in pred['year']]
    
    
    pred.to_csv(f'/home/qw64/Mt_Pleasant_GEE/pred_{this_vi}_expectile90.csv')
    obs.to_csv(f'/home/qw64/Mt_Pleasant_GEE/obs_{this_vi}_expectile90.csv')
    
    
    
```

```python
# one data frame for observed seasonal trend
for i in range(1):
    as1_obs_vi_gam=[]
    as1_obs_doy_gam=[]
    as1_obs_year_gam=[]
    as1_obs_damage_gam=[]
    # one data frame for GAM-fitted seasonal trend
    as1_pred_vi_gam=[]
    as1_pred_doy_gam=[]
    as1_pred_year_gam=[]
    as1_pred_damage_gam=[]
    
    for thisyear in ['1','0']:
        for thisdamage in ['Increase','Decrease']:
            tempdata=as1_data.query('year==@thisyear & damage==@thisdamage')
            gam_search = ExpectileGAM(expectile=0.5).gridsearch(tempdata['DOY'][:,None],tempdata['VI'])
            thislam = gam_search.lam
            gam_90 = ExpectileGAM(n_splines=8,expectile=0.9, lam=thislam).fit(tempdata['DOY'][:,None],tempdata['VI'])
            xx_pred = np.arange(1,np.max(tempdata['DOY'])+1)
            pred=gam_90.predict(xx_pred)
            xx_obs=tempdata.groupby(by='DOY').mean().index
            obs=tempdata.groupby(by='DOY').mean()
            
            as1_obs_vi_gam.extend(obs['VI'])
            as1_obs_doy_gam.extend(xx_obs)
            as1_obs_year_gam.extend(thisyear*obs.shape[0])
            as1_obs_damage_gam.extend([thisdamage]*obs.shape[0])
            
            as1_pred_vi_gam.extend(pred)
            as1_pred_doy_gam.extend(xx_pred)
            as1_pred_year_gam.extend(thisyear*len(pred))
            as1_pred_damage_gam.extend([thisdamage]*len(pred))
    
    obs=pd.DataFrame({'VI':as1_obs_vi_gam, 'DOY':as1_obs_doy_gam, 'year':as1_obs_year_gam,'damage':as1_obs_damage_gam})
    pred=pd.DataFrame({'VI':as1_pred_vi_gam, 'DOY':as1_pred_doy_gam, 'year':as1_pred_year_gam,'damage':as1_pred_damage_gam})
    
    obs['year']=['2021' if i=='1' else '2019-2020' for i in obs['year']]
    pred['year']=['2021' if i=='1' else '2019-2020' for i in pred['year']]
    
```
