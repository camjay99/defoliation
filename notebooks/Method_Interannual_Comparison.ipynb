```python
import proplot as pplt
import matplotlib as plt
import numpy as np
import pandas as pd
```

```python
turkey_point = pd.read_csv('./Data/Turkey_Point_Defoliation_Estimates.csv')
turkey_point = turkey_point.pivot(index='year', columns='method', values='sum')
```

```python
fig, ax = pplt.subplots(nrows=1, ncols=1)
ax.bar(turkey_point, cycle=['blue', 'lightblue', 'red', 'pink', 'green', 'lightgreen'], width=0.6, legend='r', legend_kw={'title':'Method','ncols':1})
ax.format(xlabel='Year', ylabel='Area ($m^2$)', title='Turkey Point', xminorticks='none', yscale='log')
fig.savefig('./Figures/Impacted_Area_Time_Series/Turkey_Point.svg')
```
