```python
import proplot as pplt
import matplotlib as plt
import numpy as np
import pandas as pd
```

```python
years = [2019, 2020, 2021, 2022, 2023]
```

## Arnot Forest

```python
arnot_forest = pd.read_csv('./Data/Arnot_Forest_Threshold_Sensitivity.csv')
arnot_forest['sum'] /= 10_000

af_thresh = [arnot_forest[(arnot_forest['year'] == year)].pivot(index='threshold', columns='method', values='sum')
                 for year in years]
af_cumul = arnot_forest.groupby(['threshold', 'method']).sum().reset_index().pivot(index='threshold', columns='method', values='sum')
```

```python
fig, ax = pplt.subplots()
ax.plot(af_thresh[2])
ax.format(xlim=[0, -0.2], xlabel='Threshold', ylabel='Area (ha)')
ax.legend(ncols=1)

fig.savefig('./Figures/Area_Sensitivity/Arnot_Forest_Area_Sensitivity_2021.svg')
```

```python
fig, axes = pplt.subplots(ncols=3, nrows=2)
for i in range(5):
    axes[i].plot(af_thresh[i])
    axes[i].format(xlim=[0, -0.2], xlabel='Threshold', ylabel='Area (ha)', title=years[i])
    
m = axes[5].plot(af_cumul) # Capture artist to make common legend
axes[5].format(xlim=[0, -0.2], xlabel='Threshold', ylabel='Area (ha)', title='Cumulative')
fig.legend(m, ncols=1, loc='r', title='Method')

fig.savefig('./Figures/Area_Sensitivity/Arnot_Forest_Area_Sensitivity_All_Years.svg')
```
